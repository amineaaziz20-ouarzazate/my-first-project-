# Exercice Pratique : Gestionnaire de Tâches Simple


  ## Objectif

  Dans cet exercice, vous allez créer un gestionnaire de tâches simple qui utilise un fichier texte pour stocker les tâches. Votre programme devra permettre à l'utilisateur d'ajouter des tâches, de les afficher et de les marquer comme terminées.

  ### Fonctionnalités à implémenter :

  1. Ajouter une nouvelle tâche au fichier.
  2. Afficher toutes les tâches existantes.
  3. Marquer une tâche comme terminée.
  4. Créer une interface utilisateur simple en ligne de commande.

  ### Instructions :

  - Utilisez un fichier nommé 'taches.txt' pour stocker les tâches.
  - Chaque ligne du fichier représente une tâche au format : "ID | Statut | Description"
  - Le statut peut être "À faire" ou "Terminée"
  

  1. Pour la fonction `ajouter_tache()` :
     - Ouvrez le fichier en mode 'a' (append) pour ajouter de nouvelles tâches.
     - Générez un ID unique pour chaque tâche (par exemple, en comptant les lignes existantes).
     - Écrivez la nouvelle tâche au format "ID | À faire | Description".

```python
def ajouter_tache(description):
    # Étape 1 : Générer l'ID
    try:
        f = open("taches.txt", "r")
        lignes = f.readlines()
        f.close()
        id_tache = len(lignes) + 1
    except:
        id_tache = 1  # Si le fichier n'existe pas

    # Étape 2 et 3 : Ajouter la tâche
    f = open("taches.txt", "a")
    f.write(str(id_tache) + " | À faire | " + description + "\n")
    f.close()

    # Étape 4 : Confirmation
    print("Tâche ajoutée avec succès !")
ajouter_tache("Préparer le rapport annuel")
```

  2. Pour la fonction `afficher_taches()` :
     - Ouvrez le fichier en mode 'r' (read).
     - Utilisez une boucle `for` pour lire et afficher chaque ligne.

```python
def afficher_taches():
    try:
        # Ouvrir le fichier en lecture
        fichier = open("taches.txt", "r")
        lignes = fichier.readlines()  # Lire toutes les lignes
        fichier.close()  # Fermer le fichier

        # Vérifier si le fichier est vide
        if len(lignes) == 0:
            print("Aucune tâche à afficher.")
        else:
            print("Liste des tâches :")
            for ligne in lignes:
                print(ligne)  # Afficher chaque ligne telle quelle

    except:
        # Si le fichier n'existe pas
        print("Le fichier n'existe pas encore.")
afficher_taches()

```

  3. Pour la fonction `marquer_terminee()` :
     - Ouvrez le fichier en mode 'r' pour lire toutes les tâches.
     - Créez une liste temporaire pour stocker les tâches modifiées.
     - Trouvez la tâche correspondant à l'ID donné et changez son statut.
     - Ouvrez le fichier en mode 'w' (write) et écrivez toutes les tâches modifiées.

```python
def marquer_terminee(id_tache):
    try:
        # Étape 1 : Lire toutes les tâches
        f = open("taches.txt", "r")
        lignes = f.readlines()
        f.close()

        # Étape 2 : Créer une liste pour stocker les tâches modifiées
        nouvelles_lignes = []

        # Étape 3 : Modifier la tâche correspondant à l'ID
        for ligne in lignes:
            # Chaque ligne a le format "ID | Statut | Description"
            parts = ligne.strip().split(" | ")
            if parts[0] == str(id_tache):  # Si l'ID correspond
                parts[1] = "Terminée"      # Changer le statut
                print(f"Tâche {id_tache} marquée comme terminée !")
            nouvelles_lignes.append(" | ".join(parts) + "\n")

        # Étape 4 : Réécrire toutes les tâches dans le fichier
        f = open("taches.txt", "w")
        f.writelines(nouvelles_lignes)
        f.close()

    except:
        print("Erreur : Le fichier n'existe pas ou l'ID est invalide.")
marquer_terminee(1)
```

```python


def menu():
      while True:
          print("\n--- Gestionnaire de tâches ---")
          print("1. Ajouter une tâche")
          print("2. Afficher les tâches")
          print("3. Marquer une tâche comme terminée")
          print("4. Quitter")
          choix = input("Choisissez une option : ")
          
          if choix == '1':
              description = input("Entrez la description de la tâche : ")
              ajouter_tache(description)
          elif choix == '2':
              afficher_taches()
          elif choix == '3':
              id_tache = input("Entrez l'ID de la tâche terminée : ")
              marquer_terminee(id_tache)
          elif choix == '4':
              print("Au revoir !")
              break
          else:
              print("Option invalide. Veuillez réessayer.")

if __name__ == "__main__":
      menu()
  
```

     - Écrivez la nouvelle tâche au format "ID | À faire | Description".

  2. Pour la fonction `afficher_taches()` :
     - Ouvrez le fichier en mode 'r' (read).
     - Utilisez une boucle `for` pour lire et afficher chaque ligne.

  3. Pour la fonction `marquer_terminee()` :
     - Ouvrez le fichier en mode 'r' pour lire toutes les tâches.

     - Trouvez la tâche correspondant à l'ID donné et changez son statut.
     - Ouvrez le fichier en mode 'w' (write) et écrivez toutes les tâches modifiées.

  4. N'oubliez pas de gérer les exceptions, par exemple si le fichier n'existe pas lors de la première utilisation.
